```python
# https://shankarmsy.github.io/posts/pca-sklearn.html
# http://www.cse.psu.edu/~rtc12/CSE486/lecture32_6pp.pdf
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('vein.jpeg') 
# Now, let's look at the size of this numpy array object img as well as plot it using imshow. 

print(img.shape) 
plt.imshow(img)
```

```python
img_r = np.reshape(img, (img.shape[0], img.shape[1] * img.shape[2])) 
print(img_r.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
ipca = PCA(32, svd_solver='randomized').fit(img_r)
```

```python
img_c = ipca.transform(img_r)
print(img_c.shape)
```

```python
compression_ratio = 100 / (100 * ((img_c.shape[0] * img_c.shape[1]) + (img_c.shape[1] * img.shape[1])) / (img.shape[0] * img.shape[1] * img.shape[2]))
represented_ratio = np.sum(ipca.explained_variance_ratio_)
print('Compression ratio: %s times' % compression_ratio)
print('Loss ratio: %s %%' % represented_ratio)
```

```python
# print(ipca.__dict__)
# print(ipca.__dict__['components_'].shape)
```

```python
temp = ipca.inverse_transform(img_c)
print(temp.shape)
temp = np.reshape(temp, img.shape)
print(temp.shape)
```

```python
plt.imshow(temp.astype(np.uint8))
```

```python
minr = 255
ming = 255
minb = 255

maxr = 0
maxg = 0
maxb = 0
for y in range(0, temp.shape[0]):
    for x in range(0, temp.shape[1]):
        if temp[y][x][0] >= maxr:
            maxr = temp[y][x][0]                
        if temp[y][x][1] >= maxg:
            maxg = temp[y][x][1]
        if temp[y][x][2] >= maxb:
            maxb = temp[y][x][2]
        
        if temp[y][x][0] <= minr:
            minr = temp[y][x][0]                
        if temp[y][x][1] <= ming:
            ming = temp[y][x][1]
        if temp[y][x][2] <= minb:
            minb = temp[y][x][2]

print(maxr, maxg, maxb)
print(minr, ming, minb)

maxc = max(maxr, maxg, maxb)
minc = max(minr, ming, minb)

y1 = 0
y2 = 255

a_r = (y2 - y1) / (maxc - minc)
a_g = (y2 - y1) / (maxc - minc)
a_b = (y2 - y1) / (maxc - minc)

b_r = (y1 * maxc - y2 * minc) / (maxc - minc)
b_g = (y1 * maxc - y2 * minc) / (maxc - minc)
b_b = (y1 * maxc - y2 * minc) / (maxc - minc)

print(a_r, a_g, a_b)
print(b_r, b_g, b_b)
```

```python
new_temp = np.zeros(temp.shape)

for y in range(0, temp.shape[0]):
    for x in range(0, temp.shape[1]):
        new_temp[y,x] = (a_r * temp[y,x,0] + b_r, a_g * temp[y,x,1] + b_g, a_b * temp[y,x,2] + b_b)
```

```python
plt.imshow(new_temp.astype(np.uint8))
```

```python
import cv2
```

```python
new_img = new_temp
size = np.size(new_img)
skel = np.zeros(new_img.shape,np.uint8)
 
ret, new_img = cv2.threshold(new_img,127,255,0)
element = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
done = False
 
while( not done):
    eroded = cv2.erode(new_img,element)
    temp_ = cv2.dilate(eroded,element)
    temp_ = cv2.subtract(new_img,temp_)
    skel = cv2.bitwise_or(skel,temp_)
    new_img = eroded.copy()
 
    zeros = size - cv2.countNonZero(new_img)
    if zeros==size:
        done = True
```
